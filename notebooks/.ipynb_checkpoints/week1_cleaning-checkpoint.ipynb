```python
import pandas as pd
import numpy as np
```

```python
import os
os.getcwd()
```

```python
# Load the CSV file
df = pd.read_csv('../data/raw/online_retail_II.csv')

print("Total rows:", df.shape[0])
print("Total columns:", df.shape[1])
print("\nColumn names:\n", df.columns.tolist())
```

```python
print(df.head())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```python
print("Rows before cleaning:", len(df))

df = df.dropna(subset=['Customer ID'])
df = df.dropna(subset=['Description'])

print("Rows after removing nulls:", len(df))
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]

print("Rows after removing cancellations:", len(df))
```

```python
df = df[df['Quantity'] > 0]
df = df[df['Price'] > 0]

print("Rows after removing bad quantity/price:", len(df))
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Customer ID'] = df['Customer ID'].astype(str)

print("Data types now:\n", df.dtypes)
```

```python
df['TotalPrice'] = df['Quantity'] * df['Price']

print(df[['Quantity', 'Price', 'TotalPrice']].head())
```

```python
print("=== FINAL CLEANED DATASET ===")
print("Total rows:", df.shape[0])
print("Unique customers:", df['Customer ID'].nunique())
print("Unique products:", df['StockCode'].nunique())
print("Unique countries:", df['Country'].nunique())
print("Date range:", df['InvoiceDate'].min(), "to", df['InvoiceDate'].max())
print("\nMissing values remaining:\n", df.isnull().sum())
```

```python
df.to_csv('../data/processed/cleaned_data.csv', index=False)

print("✅ Saved to: data/processed/cleaned_data.csv")
```
